```python
# Define URL
url = "https://www.moneycontrol.com/stocks/hist_index_result.php?indian_indices=50&pno=1&hdn=daily&fdt=2019-01-01&todt=2022-11-01"
```

```python
# Ask hosting server to fetch url
requests.get(url)
```

```python
pages = requests.get(url)
pages.text
```

```python
# parser-lxml = Change html to Python friendly format
soup = BeautifulSoup(pages.text, "lxml")
soup
```

```python
table = soup.find('table',{'tblchart'})
```

```python
headers = []
```

```python
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
```

```python
headers
```

```python
headers.remove('- SPREAD -')
```

```python
headers
```

```python
df = pd.DataFrame(columns = headers)
```

```python
df
```

```python
len(df)
```

```python
for row in table.find_all('tr')[2:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    print(length)
    df.loc[length] = row_data
```
